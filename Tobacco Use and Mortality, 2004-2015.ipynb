```python
import pandas as pd

# Correct file path
file_path = r"C:\Users\DELL\Downloads\Uber-Jan-Feb-FOIL.csv"

# Load CSV file
raw_data = pd.read_csv(file_path)

# Show the real columns inside your CSV
print("ðŸ‘‰ These are your actual column names:")
print(raw_data.columns)
print("\n")

# ----------- AUTO-DETECT FARE COLUMN ----------
fare_col = None
possible_fare_names = ['fare', 'Fare', 'Price', 'price', 'amount', 'Amount', 'fare_amount']

for col in raw_data.columns:
    if col in possible_fare_names:
        fare_col = col
        break

if fare_col:
    print(f"ðŸ‘‰ Fare column detected: {fare_col}")
    raw_data[fare_col] = (
        raw_data[fare_col]
        .astype(str)
        .str.replace('$', '', regex=False)
        .str.replace(',', '', regex=False)
    )
else:
    print("âš  No fare column found. Skipping fare cleaning.")

# ----------- AUTO-DETECT DATE COLUMN ---------
date_col = None
possible_date_names = ['date', 'Date', 'Date/Time', 'timestamp', 'datetime', 'Pickup_datetime']

for col in raw_data.columns:
    if col in possible_date_names:
        date_col = col
        break

if date_col:
    print(f"ðŸ‘‰ Date column detected: {date_col}")
    raw_data[date_col] = pd.to_datetime(raw_data[date_col], errors='coerce')
else:
    print("âš  No date column found. Skipping date conversion.")

# ----------- OUTPUT FINAL RESULT -------------
print("\nðŸ‘‰ FINAL CLEANED DATA PREVIEW:")
print(raw_data.head())
```

```python
df = pd.read_csv(r"C:\Users\DELL\Downloads\Uber-Jan-Feb-FOIL.csv")

print("Dataset loaded!")
print(df.head())
print(df.info())
```

```python
# Check missing data
print(df.isnull().sum())

# Fill missing values
df = df.fillna(method='ffill')
df = df.fillna(method='bfill')
```

```python


# Check missing data
print(df.isnull().sum())

# Forward fill
df = df.ffill()

# Backward fill
df = df.bfill()

print("Missing values after filling:")
print(df.isnull().sum())
```

```python
# Statistical summary
print(df.describe())

# Histogram
df.hist(figsize=(18, 12))
plt.show()

# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=['int64', 'float64'])

plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap (Numeric Features Only)")
plt.show()

```
